```python
import matplotlib.pyplot as plt

def solve_txt(file_path) :
    with open(file_path, 'r') as f:
        li = f.read()
    f.close()
    li = li.replace(',',';')
    li = li.replace(' ','')

    li = li.split('\n')[1:-1]
    li = [i.split(';') for i in li]
    for j in range(len(li)) :
        for i in range(len(li[j])) :
            if li[j][i] :
                s = li[j][i].split(':')
                li[j][i] =  {s[0] : float(s[1])}   
        li[j] = {**li[j][0],**li[j][1],**li[j][2]}
    final = []
    for i in range(0, len(li)-1, 2) :
        final.append({**li[i],**li[i+1]})
   
    return final

def plot_result(feature, data1, data2,  label1, label2, title) :
    fig = plt.figure(feature, figsize=(10,8))
    ax1 = fig.add_subplot()
    ax2 = fig.add_subplot()
    args_ax1 = [range(len(data1)), [i[feature] for i in data1]]
    args_ax2 = [range(len(data2)), [i[feature] for i in data2]]

    config_plot = dict(
        linestyle='solid',
        linewidth='2',
        marker='o',
        markersize=1,
        markeredgecolor='white',
        markeredgewidth=3,
    )

    ax1.plot(label=label1, *args_ax1, **config_plot)
    ax2.plot(label=label2, *args_ax2, **config_plot)
    for x in [0,20,30,40] :
        ax1.axvline(x, color='black', linestyle='dashed')

    plt.title(f"{title}", fontsize=20)
    ax1.legend(loc='upper right', fontsize=15, shadow=True) ## 범례 생성 - 그림자 효과 추가
    ax2.legend(loc='upper right', fontsize=15, shadow=True)
```

```python
dir = './logs'
file_names = ['ex2-1.txt', 'ex2-2.txt']
result=[]
for file in file_names :
    result.append(solve_txt(dir+'/'+file))
```

```python
features = ['train_loss', 'val_loss','score', 'score_g', 'score_v', 'score_c']
for feature in features :
    plot_result(feature, result[0], result[1],  'withcut', 'withoutcut', f'{feature}_randomRotation_comparision')
```
